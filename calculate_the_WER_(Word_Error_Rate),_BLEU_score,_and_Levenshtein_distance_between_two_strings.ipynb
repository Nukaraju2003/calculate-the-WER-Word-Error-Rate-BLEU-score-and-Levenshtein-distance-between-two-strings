```python
import nltk
from nltk.translate.bleu_score import sentence_bleu

# Example usage:
texttransducer = "hi my name is julie saha i was born and raised in m p japan i have completed graduation in commerce in the year two thousand sixteen with correspondence we are a family of five my mother in law my husband my two kids and i i uh i have overall eight and a half years of work experience for the international b p o i have worked in inbound and outbound customer care and semi technical process in my leisure time i like to spend most of my time reading books listening to music travelling and of course and with my kids about uh talking about my strength i would say i'm self motivated i do not need someone else to push me to achieve my goals and dreams and about weakness i would say um i look for perfection i completely understand it is humanly not always possible to be perfect hence i'm working on uh my weakness to make sure that i figure out a way to get things done and in a better way even if it is not perfect and um yes uh that's mostly about me thank you so much for giving me this opportunity to talk about myself and introduce myself to you all hope to hear from you soon thank you you have a good day bye"
textgoldstandard = "Hi, my name is Julie Saha. I was born and raised in MP, Jodhpur. I have completed graduation in commerce in the year 2016 with correspondence. We are a family of five, my mother-in-law, my husband, my two kids and I. I have overall eight and a half years of work experience with international BPO. I have worked in inbound and outbound, customer care and semi-technical processes. In my leisure time, I like to spend most of my time reading books, listening to music, traveling and of course spending time with my kids. About talking about my strength, I would say I'm self-motivated. I do not need someone else to push me to achieve my goals and dreams. And about weakness, I would say I look for perfection. I completely understand it is humanly not always possible to be perfect. Hence, I'm working on my weakness to make sure that I figure out a way to get things done in a better way even if it is not perfect. And yes, that's mostly about me. Thank you so much for giving me this opportunity to talk about myself and introduce myself to you all. Hope to hear from you soon. Thank you. You have a good day. Bye.Bye."

hypothesis = texttransducer
reference = textgoldstandard

def calculate_wer(hypothesis, reference):
    hypothesis_words = hypothesis.split()
    reference_words = reference.split()

    distance = nltk.edit_distance(hypothesis_words, reference_words)
    wer = distance / len(reference_words)
    return wer

def calculate_bleu(hypothesis, reference):
    hypothesis_words = hypothesis.split()
    reference_words = reference.split()

    weights = tuple([1.0 / len(reference_words)] * len(reference_words))
    bleu = sentence_bleu([reference_words], hypothesis_words, weights=weights)
    return bleu

def calculate_levenshtein(hypothesis, reference):
    distance = nltk.edit_distance(hypothesis, reference)
    return distance

wer = calculate_wer(hypothesis, reference)
bleu = calculate_bleu(hypothesis, reference)
levenshtein = calculate_levenshtein(hypothesis, reference)

print("WER:", wer)
print("BLEU:", bleu)
print("Levenshtein distance:", levenshtein)
```
